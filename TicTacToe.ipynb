# Tic Tac Toe game

# Enter numbers 1-9 to play the game.

# Take turns between players.

# Player 1 is X and Player 2 is Y 

# This game uses standard computer keypad format

```python
import math
import time
import numpy as np
from IPython.display import clear_output

s = np.empty((3, 3), dtype='str')
s[:] = ' '
moves = []



def board():
    global s
    clear_output()

    print('   |   |')
    print(' '  + s[0][0] + ' | ' + s[0][1] + ' | ' + s[0][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' '  + s[1][0] + ' | ' + s[1][1] + ' | ' + s[1][2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' '  + s[2][0] + ' | ' + s[2][1] + ' | ' + s[2][2])
    print('   |   |')
        
def available(g):
    global s
    r = (9-g)/3 + 1
    c = g + (3*r) - 9
    if s[r-1][c-1]== 'X' or s[r-1][c-1] == 'Y':
        return True
    else:
        return False
    

def win_combo(h):
    
    global s
    
    if s[0][0]== h and s[0][1] == h and s[0][2]== h: #TOP
        return True
    elif s[1][0] == h and s[1][1] == h and s[1][2] == h: #MIDDLE
        return True 
    elif s[2][0] == h and s[2][1] == h and s[2][2] == h: #BOTTOM
        return True 
    elif s[0][0] == h and s[1][0] == h and s[2][0] == h: #DOWN LEFT
        return True 
    elif s[0][1] == h and s[1][1] == h and s[2][1] == h: #DOWN MIDDLE
        return True
    elif s[0][2] == h and s[1][2] == h and s[2][2] == h: #DOWN RIGHT
        return True
    elif s[0][0] == h and s[1][1] == h and s[2][2] == h: #DIAGONAL 1
        return True
    elif s[0][2] == h and s[1][1] == h and s[2][0] == h: #DIAGONAL 2
        return True
    else:
        return False
    
        
def drawed(num):
    
    global s
    global moves
    Checklist = [i for i in range(1,10)]
    moves.append(num)
    moves.sort()
    if moves == Checklist:
        return True
    else:
        return False
    
    
def game_on():
    
    global s
    global moves
    print "Let's play TIC TAC TOE\n"
    #new_game()
    del moves[:]
    s[:] = ' '
    board()
    print ''
    ch = True
    while(1): 
        print 'Player 1'
        print ''
        
        while True:
            play1 = int(raw_input("Please enter your choice : "))
            if play1 not in range(1,10) or available(play1):
                print ''
                print 'Invalid Choice...Please re-enter'
                continue
            else:
                break
        print ''
        
        row1 = (9-play1)/3 + 1
        col1 = play1 + (3*row1) - 9
        s[row1-1][col1-1] = 'X'
        board()
        
        if win_combo('X'):
            print ""
            print "Congrats !! Player 1 is the winner !!"
            print ""
            break
            
        if drawed(play1):
            print ''
            print "It is a draw !"
            print ""
            break
            
        print ''
        print 'Player 2'
        print ''
        
        while True:
            play2 = int(raw_input("Please enter your choice : "))
            if play2 not in range(1,10) or available(play2):
                print ''
                print 'Invalid Choice...Please re-enter'
                continue
            else:
                break
                

        print''
        
        row2 = (9-play2)/3 + 1
        col2 = play2 + (3*row2) - 9
        s[row2-1][col2-1] = 'Y'
        board()
        
        if win_combo('Y'):
            print ""
            print "Congrats !! Player 2 is the winner !!"
            print ""
            break
            
        if drawed(play2):
            print "It is a draw !"
            print ""
            break
            
        print''
    print "\n\n"
    print "Let's play again!"
    print ""
    time.sleep(5)
    clear_output()
    game_on()
```
